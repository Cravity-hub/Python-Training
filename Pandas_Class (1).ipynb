### Combining Pandas Series in to a DataFrame

There are several ways to concatenate two series in pandas. Some of them are higlighted below.

```python
#first we'll create three data series

def create_series(series_data):
    """create pandas series"""
    
    pd_series = pd.Series(series_data)
    
    return pd_series

employee_name = create_series(['John Smith', 'Ola Badmus', 'Susan Patrick', 
                               'Musa Ibrahim', 'Ada Chukwuma'])

gender = create_series(['Male', 'Male', 'Female', 'Male', 'Female'])


salary = create_series([50000, 65000, 80000, 50000, 60000])


#then we create a dictionary of our series

data = {'employee_name': employee_name,
       'gender': gender,
       'salary': salary}

#finally we'll use the concat method to combine the series into a datafame object
df = pd.concat(data, axis = 1)
df
```

```python
stocks = create_series(['APPLE', 'TESLA', 'MICROSOFT', 'S&P 500'])

more_stocks = create_series(['COCA COLA', 'OANDO'])

stocks_append = stocks.append(more_stocks, ignore_index = True)

df2 = pd.DataFrame(stocks_append, columns = ["stocks"])

df2
```

### Labels
If nothing else is specified, the values are labeled with their index number. First value has index 0, second value has index 1 etc.

This label can be used to access a specified value.

### Create Labels
With the index argument, you can name your own labels.

```python
a = [1, 7, 2,4,5, 1, 0, 6]

myvar_2 = pd.Series(a, index = ["s", "t", "u","v", "w", "x", "y", "Z"])

myvar_2
```

When you have created labels, you can access an item by referring to the label.

```python
print(myvar_2["t"])
```

### Key/Value Objects as Series
You can also use a key/value object, like a dictionary, when creating a Series.

Note: The keys of the dictionary become the labels.

To select only some of the items in the dictionary, use the index argument and specify only the items you want to include in the Series.



```python
calories = {"day1": 420, "day2": 380, "day3": 390, "day4": 410}

myvar = pd.Series(calories, index = ["day1", "day2", "day4"])

myvar
```

### Renaming columns in Pandas

### Data Cleaning and Data Type Conversion

Our data has numeric, date and categorical columns but they are being shown as object type. 

We need to convert these columns to the appropriate data types. 

### Removing Regular Expressions from columns

```python
data['Budget'].str.replace(r'\D+', '', regex=True)
data['Box_Office_Rev'].str.replace(r'\D+', '', regex=True)
data['Profit'].str.replace(r'\D+', '', regex=True)

data.head()
```

```python
data['Budget'] = pd.to_numeric(data['Budget'], errors='coerce')
data['Budget'] = data['Budget'].astype(int)
```

```python
data['Box_Office_Rev'] = pd.to_numeric(data['Box_Office_Rev'])
data['Box_Office_Rev'] = data['Box_Office_Rev'].astype(int)
```

```python
data['Profit'] = pd.to_numeric(data['Profit'], errors='coerce')
data['Profit'] = data['Profit'].astype(int)
```

### Convert Release date to date type

```python
data['Release_Date'] = pd.to_datetime(data['Release_Date'], format='%d/%m/%Y')
```

```python
data['Genre'] = data.Genre.astype('category')
```

## Using Faker, Random and Pandas to create a dummy Dataset.

```python
from faker import Faker
import random
import pandas as pd
Faker.seed(0)
random.seed(0)
fake = Faker("de_DE") 
name, city_name, country, job, age = [[] for k in range(0,5)] 
for row in range(0,100):
    name.append(fake.name())
    city_name.append(fake.city_name())
    country.append(fake.country())
    job.append(fake.job())
    age.append(random.randint(20,100))

d = {"Name":name, "Age":age, "City":city_name, "Country":country, "Job":job}
dfake = pd.DataFrame(d)
dfake.head()
```
