```python
df1 = pd.DataFrame([['A', 4,6,8], ['B', 9,3,5], ['C', 2,5,8]], 
                   columns = ['grade', 'maths', 'english', 'physics'])
df1
```

```python
df2 = pd.DataFrame([['A', 4,6,8], ['B', 9,3,5], ['D', 4,6,3], ['E', 6,8,9]], 
                   columns = ['grade', 'maths', 'english', 'government'])
df2
```

```python
df1.merge(df2, how = 'inner', on = 'grade', suffixes = ['_sci', '_arts'])
```

```python
df1.merge(df2, how = 'outer', on = 'grade', suffixes = ['_sci', '_arts'])
```

```python
df1.merge(df2, how = 'left', on = 'grade', suffixes = ['_sci', '_arts'])
```

```python
df1.merge(df2, how = 'right', on = 'grade', suffixes = ['_sci', '_arts'])
```

```python
df1.merge(df2, how = 'inner', left_on = 'physics', right_on = 'government', suffixes = ['_sci', '_arts'])
```
