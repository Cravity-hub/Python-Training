



## What is Statistics?

Statistics is an applied science concerned with analysis and modeling of data. Statistical analysis is the process of collecting and analyzing data in order to discern patterns and trends. 

It is a method for removing bias from evaluating data by employing numerical analysis. 

This technique is useful for collecting the interpretations of research, developing statistical models, and planning surveys and studies.

## Types of Data

Data comes from many sources: sensor measurements, events, text, images and videos. Data can be structured or unstructured. To apply statitstical concepts to unstructured data, it must be processed into structured form. 

### Elements of Structured Data
There are two types of structured data: numeric and categorical. It suffice to say that for the purposes
of data analysis and predictive modeling, the data type is important to help determine the type of visual display, data analysis, or statistical model. 

* **Continuous**
Data that can take on any value in an interval. They are also known as interval, float or numeric data types.

* **Discrete**
Data that can take on only integer values, such as counts. They can also be called integers or count data types.

* **Categorical**
Data that can take on only a specific set of values representing a set of possible categories. They can also be called nominal, polychotomous.

* **Binary**
A special case of catgorical data with just two categories of values (0/1, True/False)

* **Ordinal**
Categorical data that has an explicit ordering. 

### Rectangular Data
The typical frame of reference for an analysis in data science is a rectangular data object, like a spreadsheet or database table.

* **Data Frame**
Rectangular data (like a spreadsheet) is the basic data structure for statistical and ML models.

* **Feature**
A column in the table is commonly referred to as a feature. Other names are attribute, input, predictor, variable.


* **Outcome**
The features are sometimes used to predict the outcome in an experiment or study. Many data science projects involve predicting an outcome. Features can also be called dependent variable, response, target, or output.

* **Records**
A row in the table is commonly referred to as a record. Other names for records are case, example, instance, observation, pattern and sample.

* **Outliers**
An outlier is any value that is very distant from the other values in a data set.

## Exploratory Data Analysis. 


Exploratory Data Analysis (EDA) is an approach used by data scientists to analyze and investigate data sets and summarize their main characteristics, often employing data visualization methods. 

It is used to analyze the data and discover trends, patterns, or check assumptions in data with the help of statistical summaries and graphical representations. 

The goal of EDA is to gain insights into the data set that can be used for further analysis or modeling.

### Descriptive Statistics

This aspect of statistics seeks to summarize data. In descriptive statistics, we calculate values that give us a summary of our dataset. These values are called statistic. DS has two parts:

* Measures of Central Tendency
* Measures of Dispersion

#### * Measures of Central Tendency
This involves computing one of several different types of averages to give you the one best data point that represents a set of data.

Averages, also called measures of central tendency, come in three flavors: the mean, the
median, and the mode. Each provides you with a different type of information about a
distribution of scores and is simple to compute and interpret.


#### The Mean

* The mean is the most common type of average that is computed. It is simply the sum of all the values in a group, divided by the number of values in that group.

* The mean is like the fulcrum on a seesaw. Its the centermost point where all the values on one side of the mean are equal in weight to all the values on the other side of the mean.

* The mean is also sensitive to extreme scores. Extreme scores are called outliers. 

##### Types of Mean

1. Weighted Mean

All the values in the data are assigned weights. 


2. Trimmed Mean

This mean is computed by trimming both ends of the data. This method of computing the mean removes the influence of outliers. 

#### The Median

* The median refers to the value with position at the cenrtre to the data. The position of the values in the data rather than the values themselves are considered here. 

* Unlike the mean, the median is less sensitive to outliers. However, it does not capture all the values in the data. 

* The basic metric for location is the mean, but it can be sensitive to extreme values (outlier). Other metrics (median, trimmed mean) are more robust.

#### The Mode

* The is the value with the most frequency.


### Measures of Dispersion

These statistics has to do with measuring how spread out you data is around the cetral tendencies. They help us to know to what extent our averages represent the data. 

* Range
* Quartiles & Percentiles
* Interquartile Range
* Variance
* Standard Deviation


#### Range

Statistically Range refers to the difference beteween the maximum and the minimum values in the data. 

#### Quartiles & Percentiles

#### IQR
This is the difference between the third quartile and the first quartile.

#### Variance
The Variance measures the average distance between the mean and other values in the data. The downside of the variance is that it is not in the same unit as the quantity being measured because of the squares.

#### Standard Deviation

Is calculated by taking the squareroot of the variance. And it is measured in the same unit as the quantity being studied. 

## Inferential Statistics
Inferential statistics is a branch of statistics that uses sample data to infer properties or make conclusions about a larger population. It is based on the assumption that the sample data is representative of the population data. Inferential statistics is used to test hypotheses, derive estimates, and make predictions. It is more cost-effective and less tedious than collecting data from an entire population.



For example, you can use inferential statistics to estimate the mean UTME score of all final year O'level students in Nigeria based on a sample of 1000 students.


The goal of inferential statistics is to make generalizations about the population. In inferential statistics, a statistic is taken from the sample data (e.g., the sample mean) that used to make inferences about the population parameter (e.g., the population mean).


### Types of Inferential Statistics

Inferential statistics can be classified into hypothesis testing and regression analysis. 

#### Hypothesis testing

Is a type of inferential statistics that is used to test assumptions and draw conclusions about the population from the sample data. 


#### Z Test
The Z Test is used on a data that follows a normal distribution and has a sample size greater than or equal to 30. it is used to test if the means of the sample and population are equal when the population variance is known. 


#### T Test
A t test is used when the data follows a t distribution and the sample size is lesser than 30. it is used to compare the sample and population mean when the population variance is unknown. 


#### F Test
An F Test is used to check if there is a difference between the variance of two sample or populations.


#### Correlation Test 
Is used to test for the significance of the relationship between two numerical variables. 


#### Confidence Interval
A confidence interval helps in estimating the parameters of a population. For instance, a 95% confidence interval tells us that if a test is conducted 100 times with new samples every time, you can expect the estimate to lie within the given interval 95 times. 


#### Regression Analysis
Regression analysis is used to quantify how one variable will change wrt another variable. There are many types of regressions available such as simple linear, multiple linear, nominal, logistic, and ordinal regression.

#### To perform a one-way ANOVA on origin and mpg

##### Research question
Is there a difference between the mean mpg of cars based on their origins?

##### Hypothesis

Ho: There is no signicant difference in the mean mpg of cars by origin. 


H1: There is a significant difference in the mean mpg of cars by origin. 


alpha-value = 0.05 (5%)




```python
# import the necessary module
from scipy.stats import f_oneway

# create  the groups from the categorical variable (origin)
europe = df['mpg'][df['origin'] == 'europe']
japan = df['mpg'][df['origin'] == 'japan']
usa = df['mpg'][df['origin'] == 'usa']


statistic, pvalue = f_oneway(europe, japan, usa)

print('The statistic is: ', statistic)
print('The pvalue is: ', pvalue)


if pvalue < 0.05:
    print('Reject H0')
else:
    print('We fail to reject H0')
```
